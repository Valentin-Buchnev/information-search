```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('compression.csv', sep=',', names=['p', 'compressionRatio'])
```

---
Отрисуем зависимость степени сжатия от $p$.

```python
plt.figure(figsize=(15, 8))
plt.plot(data['p'], data['compressionRatio'], label='compresssion ratio')
plt.legend(fontsize=16)
plt.grid()
plt.xlabel('p from geometric distribution', fontsize=16)
plt.ylabel('ratio', fontsize=16)
plt.show()
```

---
И еще раз, с логарифмическими осями.

```python
plt.figure(figsize=(15, 8))
plt.plot(data['p'], data['compressionRatio'], label='compresssion ratio')
plt.legend(fontsize=16)
plt.grid()
plt.xlabel('p from geometric distribution', fontsize=16)
plt.ylabel('ratio', fontsize=16)
plt.yscale('log')
plt.xscale('log')
plt.show()
```
