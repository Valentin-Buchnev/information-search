```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('result.csv', sep=',', names=['id', 'groupSize'])
```

```python
plt.figure(figsize=(15, 8))

plt.hist(data['id'], weights=data['groupSize'], bins=500, log=True)
plt.yscale('log')
plt.xscale('log')
plt.title('Histogram of groups distribution by their size', fontsize=24)
plt.xlabel('size of group', fontsize=16)
plt.ylabel('groups count', fontsize=16)
plt.show()
```

---
### Алгоритм

Использовалось разделение хэша на 4 части (каждая по 16 бит) и тот факт, что, если хэши различаются не более чем в 3 битах, то найдется часть, где эти хэши совпадают.

Далее, при обработке очередного хэша $h^{(i)}$, перебираем неизменяющуюся часть хэша $h_{k}^{(i)}$ ($k \in \{1, 2, 3, 4\}$) и все хеши $h^{(j)}$, имеющие такую же часть $h_{k}^{(j)}$ и считаем, сколько из них имеют не более 3-х различающихся бит.
