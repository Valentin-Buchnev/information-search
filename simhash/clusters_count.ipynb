```python
import requests
import json
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook as tqdm
import urllib.parse
import re
from matplotlib import pyplot as plt
import csv
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
clusters_count = pd.read_csv('out_map_full.txt', sep='\t', header=None, names=['size', 'count'])
```

```python
plt.figure(figsize=(15, 6))
plt.title('Count of clusters by size', fontsize=16)
plt.xlabel('Cluster size', fontsize=14)
plt.ylabel('Clusters count', fontsize=14)
plt.grid()
plt.hist(clusters_count['size'], weights=clusters_count['count'], 
         bins=[i for i in range(500)], log=True)
plt.yscale('log')
plt.xscale('log')
plt.xticks([2 ** i for i in range(10)], labels=[2 ** i for i in range(10)])
plt.show()
```
